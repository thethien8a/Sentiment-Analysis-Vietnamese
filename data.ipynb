```python
from googleapiclient.discovery import build
import pandas as pd
import time
```

```python
api_key = 'AIzaSyBBr0fKAthXn_HzRgfNHVXsf5u6Odo-aU4'
```

```python
youtube = build(
    'youtube',
    'v3',
    developerKey = api_key
)
comments_data = []
next_page_token = None
vid_id ='iK-Cji6J73Q'
```

```python
while True:
    request = youtube.commentThreads().list(
        part = 'snippet',
        videoId = vid_id,
        order = 'time',
        maxResults=100,
        textFormat="plainText",
        pageToken=next_page_token
    )

    response = request.execute()

    for item in response["items"]:
        comment_snippet = item["snippet"]["topLevelComment"]["snippet"]

        comment_info = {
            "authorDisplayName": comment_snippet.get("authorDisplayName"),
            "videoId": vid_id,
            "textDisplay": comment_snippet.get("textDisplay"),
            "likeCount": comment_snippet.get("likeCount"),
            "publishedAt": comment_snippet.get("publishedAt"),
            "totalReplyCount": item["snippet"].get("totalReplyCount"),
        }

        comments_data.append(comment_info)

    next_page_token = response.get("nextPageToken")
    if not next_page_token:
        break

    df = pd.DataFrame(comments_data)
```
